### Perform Data Cleaning [ Replace '?' with NaN ]

```python
df.isin(['?']).sum()
```

```python
# replace ? with NaN
df["workclass"] = df["workclass"].replace('?', np.nan)
df["occupation"] = df["occupation"].replace('?', np.nan)
df["native-country"] = df["native-country"].replace('?', np.nan)

df.isin(['?']).sum()
```

```python
df.isna().sum()
```

```python
sns.heatmap(df.isna())  
```

```python
df.dropna(inplace=True)
```

```python
print(df.isna().sum())
sns.heatmap(df.isna())  
```

### Check for duplicates and drop them

```python
dup = df.duplicated()
dup.sum()  # total of 47 duplicated rows
```

```python
# we can keep the first/ last occurrence of the duplicated rows or keep none
# also we can drop the duplicates based on a subset of columns  --> 
# df_no_duplicates_city = df.drop_duplicates(subset=['City'])

df = df.drop_duplicates(keep='first')
df.shape
```
